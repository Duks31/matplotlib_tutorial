```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

```python
xpoints = np.array([1,2,3,4,5,6])
ypoints = np.array([3,6,9,12,15,18])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
ypoints = np.array([3,1,9,2,15,8])
plt.plot(ypoints)
plt.show()
```

```python
xpoints = np.array([3,10,8,1])
ypoints = np.array([2,5,7,11])
plt.plot(xpoints,ypoints,marker = 'o')
plt.show()
```

```python
xpoints = np.array([4,8,7,1])
ypoints = np.array([3,6,11,9])
plt.plot(xpoints,ypoints,'*')
plt.show()
```

```python
xpoints = np.array([3,10,8,1])
ypoints = np.array([2,5,7,11])
plt.plot(xpoints,ypoints,'o--g')
plt.show()
```

```python
ypoints = np.array([3,1,9,2,15,8])
plt.plot(ypoints,marker = 'o',ms = '15')
plt.show()
```

```python
ypoints = np.array([3,1,9,2,15,8])
plt.plot(ypoints,marker = 'o',ms = '15',mec = 'r')
plt.show()
```
